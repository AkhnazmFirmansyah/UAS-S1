**Soal 5: Program Manajemen Kursus dengan PBO, Struktur Data, Fungsi, dan Penanganan Kesalahan**

```python
class Kursus:
    def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
        self.nama_kursus = nama_kursus
        self.kode_kursus = kode_kursus
        self.durasi = durasi
        self.kapasitas = kapasitas
        self.peserta = 0

    def tambah_peserta(self):
        """Menambah peserta kursus jika kapasitas belum penuh."""
        if self.peserta < self.kapasitas:
            self.peserta += 1
            print("Peserta berhasil ditambahkan ke kursus" ,self.nama_kursus)
        else:
            print("Error: Kapasitas kursus", self.nama_kursus, "sudah penuh.")

    def info_kursus(self):
        """Menampilkan informasi lengkap tentang kursus."""
        return (f"Nama Kursus: {self.nama_kursus}\n"
                f"Kode Kursus: {self.kode_kursus}\n"
                f"Durasi: {self.durasi} minggu\n"
                f"Kapasitas: {self.kapasitas}\n"
                f"Peserta Terdaftar: {self.peserta}\n")


daftar_kursus = []
kursus_dict = {}

def tampilkan_semua_kursus():
    """Menampilkan semua kursus yang tersedia."""
    if daftar_kursus:
        for kursus in daftar_kursus:
            print(kursus.info_kursus())
            print("-" * 40)
    else:
        print("Tidak ada kursus yang tersedia.")

def cari_kursus(kode):
    """Mencari kursus berdasarkan kode dan menampilkan informasi lengkap."""
    if kode in kursus_dict:
        print(kursus_dict[kode].info_kursus())
    else:
        print("Error: Kursus dengan kode", kode, "tidak ditemukan.")

def tambah_kursus():
    """Menambahkan kursus baru."""
    try:
        nama_kursus = input("Masukkan nama kursus: ")
        kode_kursus = input("Masukkan kode kursus: ")
        durasi = int(input("Masukkan durasi kursus (dalam minggu): "))
        kapasitas = int(input("Masukkan kapasitas kursus: "))

        kursus_baru = Kursus(nama_kursus, kode_kursus, durasi, kapasitas)

        daftar_kursus.append(kursus_baru)
        kursus_dict[kode_kursus] = kursus_baru

        print("Kursus ",nama_kursus, "dengan kode", kode_kursus, "berhasil ditambahkan.")
    except ValueError:
        print("Error: Durasi dan kapasitas harus berupa angka.")

def tambah_peserta_ke_kursus():
    """Menambahkan peserta ke kursus tertentu."""
    kode_kursus = input("Masukkan kode kursus untuk menambah peserta: ")
    if kode_kursus in kursus_dict:
        kursus = kursus_dict[kode_kursus]
        kursus.tambah_peserta()
    else:
        print("Error: Kursus dengan kode", kode_kursus, "tidak ditemukan.")

def menu():
    while True:
        print("\n--- Menu Manajemen Kursus ---")
        print("1. Tambah Kursus Baru")
        print("2. Tambah Peserta ke Kursus")
        print("3. Cari Kursus berdasarkan Kode")
        print("4. Tampilkan Semua Kursus")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tambah_kursus()
        elif pilihan == '2':
            tambah_peserta_ke_kursus()
        elif pilihan == '3':
            kode_kursus = input("Masukkan kode kursus yang ingin dicari: ")
            cari_kursus(kode_kursus)
        elif pilihan == '4':
            tampilkan_semua_kursus()
        elif pilihan == '5':
            print("Terima kasih telah menggunakan program manajemen kursus!")
            break
        else:
            print("Pilihan tidak valid. Silakan pilih antara 1 hingga 5.")

if __name__ == "__main__":
    menu()
```
